```python
from scipy.stats import norm
```

```python
norm.cdf(1.96) # Calculate the cumulative distribution function for a standard normal distribution at z = 1.96
```

```python
#Faire 1m77 ou moins quand la moyenne est 1m78 et l'écart type est de 6cm
print(norm.cdf(177, 178, 6)) # Calculate the cumulative distribution function for a normal distribution with mean 178 and standard deviation 6 at x = 177
#Faire 1m77 ou plus quand la moyenne est 1m78 et l'écart type est de 6cm
print(norm.sf(177, 178, 6))  # Calculate the survival function (1 - CDF) for the same normal distribution at x = 177
```

```python
norm.cdf(177, 178, 6) - norm.cdf(172, 178, 6) # Calculate the probability that a value from the normal distribution with mean 178 and standard deviation 6 falls between 172 and 177
```

```python
#Quelle taille sont les 10% des personnes les plus grandes dans une population avec une moyenne de 178cm et un écart type de 6cm?
norm.ppf(0.9, 178, 6) # Find the value x such that P(X <= x) = 0.9 for a normal distribution with mean 178 and standard deviation 6
#Quelle taille sont les 5% des personnes les plus petites dans une population avec une moyenne de 178cm et un écart type de 6cm?
norm.ppf((1-0.95), 178, 6) # Find the value x such that P(X <= x) = (1 - 0.95)/2 for a normal distribution with mean 178 and standard deviation 6
```

```python
import numpy as np
from matplotlib import pyplot as plt

# Paramètres de la loi normale standard
mean = 0
std = 1

# Créer un range de valeurs x
x = np.linspace(-4, 4, 1000)

# Calculer les valeurs de la fonction de densité de probabilité
y = norm.pdf(x, loc=mean, scale=std)

# Créer le graphique
plt.figure(figsize=(12, 7))
plt.plot(x, y, color='black', linewidth=2, label='Densité normale')

# Lignes repères de la règle 68-95-99,7
intervals = [
    (1, 'green', '68 % : ±1σ'),
    (2, 'royalblue', '95 % : ±2σ'),
    (3, 'darkorange', '99,7 % : ±3σ'),
]

for sigma, color, label in intervals:
    plt.axvline(mean - sigma * std, color=color, linestyle='--', linewidth=2, label=label)
    plt.axvline(mean + sigma * std, color=color, linestyle='--', linewidth=2)

plt.axvline(mean, color='crimson', linestyle='-', linewidth=2, label='Moyenne μ')
plt.title('Loi normale standard et règle 68-95-99,7')
plt.xlabel('x')
plt.ylabel('Densité de probabilité')
plt.xticks(np.arange(-4, 5, 1))
plt.grid(True, alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```
