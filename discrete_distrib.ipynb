## Visualisation d'une loi binomiale

La loi binomiale modélise le nombre de succès obtenus sur un nombre fixe d'essais indépendants, lorsque la probabilité de succès reste constante à chaque essai.

Dans cet exemple :

- on effectue $n = 10$ essais ;
- la probabilité de succès à chaque essai est $p = 0{,}4$ ;
- la variable aléatoire $X$ représente le nombre total de succès.

Le graphique ci-dessous montre la probabilité d'obtenir exactement $k$ succès pour chaque valeur possible de $k$. On met aussi en évidence le cas $k = 7$, déjà calculé plus haut.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.4
k_highlight = 7

k_values = np.arange(0, n + 1)
pmf_values = binom.pmf(k_values, n=n, p=p)
expected_value = n * p

colors = ['#9ecae1' if k != k_highlight else '#ef3b2c' for k in k_values]

plt.figure(figsize=(12, 6))
plt.bar(k_values, pmf_values, color=colors, edgecolor='black', width=0.75)
plt.plot(k_values, pmf_values, color='#08519c', linewidth=2, marker='o')

plt.axvline(expected_value, color='#2ca25f', linestyle='--', linewidth=2, label=f'Espérance E(X) = np = {expected_value:.1f}')
plt.axvline(k_highlight, color='#ef3b2c', linestyle=':', linewidth=2, label=f'Cas étudié : k = {k_highlight}')

highlight_prob = binom.pmf(k_highlight, n=n, p=p)
plt.annotate(
    f'P(X = {k_highlight}) = {highlight_prob:.3f}',
    xy=(k_highlight, highlight_prob),
    xytext=(k_highlight - 2.4, highlight_prob + 0.05),
    arrowprops=dict(arrowstyle='->', color='#ef3b2c', lw=1.5),
    fontsize=11,
    color='#7f0000'
 )

plt.title('Loi binomiale : probabilité du nombre de succès sur 10 essais')
plt.xlabel('Nombre de succès')
plt.ylabel('Probabilité P(X = k)')
plt.xticks(k_values)
plt.ylim(0, max(pmf_values) + 0.08)
plt.grid(axis='y', alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```
