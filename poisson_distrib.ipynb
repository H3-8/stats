```python
from scipy.stats import poisson
import numpy as np
```

```python
# If the average number of adoptions per week is 8, what is the probability of having 5 adoptions in a week?
lambda_ = 8  # average number of adoptions per week
k = 5  # number of adoptions we want to find the probability for
probability = poisson.pmf(k, lambda_)
print(f"The probability of having {k} adoptions in a week is: {probability:.4f}")
```

```python
probability = poisson.cdf(k, lambda_)
print(f"The probability of having at most {k} adoptions in a week is: {probability:.4f}")
```

```python
probability = poisson.sf(k, lambda_)
print(f"The probability of having more than {k} adoptions in a week is: {probability:.4f}")
```

## Visualisation de la loi de Poisson

La loi de Poisson modélise le nombre d'événements observés sur un intervalle fixe lorsque ces événements arrivent de façon indépendante avec un rythme moyen constant $\lambda$.

Dans cet exemple, on prend $\lambda = 8$ adoptions par semaine. Le graphique ci-dessous montre :

- la probabilité d'observer exactement $k$ adoptions pour différentes valeurs de $k$ ;
- la moyenne attendue, égale à $\lambda$ ;
- une mise en évidence du cas $k = 5$ pour relier le calcul précédent au graphique.

Plus une barre est haute, plus ce nombre d'adoptions est probable.

```python
import matplotlib.pyplot as plt

lambda_ = 8
k_highlight = 5

# Choix d'une plage couvrant l'essentiel de la masse de probabilité
k_values = np.arange(0, 21)
pmf_values = poisson.pmf(k_values, lambda_)

colors = ['#9ecae1' if k != k_highlight else '#ef3b2c' for k in k_values]

plt.figure(figsize=(12, 6))
plt.bar(k_values, pmf_values, color=colors, edgecolor='black', width=0.8)
plt.plot(k_values, pmf_values, color='#08519c', linewidth=2, marker='o')

plt.axvline(lambda_, color='#2ca25f', linestyle='--', linewidth=2, label=f'Moyenne λ = {lambda_}')
plt.axvline(k_highlight, color='#ef3b2c', linestyle=':', linewidth=2, label=f'Cas étudié : k = {k_highlight}')

highlight_prob = poisson.pmf(k_highlight, lambda_)
plt.annotate(
    f'P(X = {k_highlight}) = {highlight_prob:.3f}',
    xy=(k_highlight, highlight_prob),
    xytext=(k_highlight + 1.2, highlight_prob + 0.03),
    arrowprops=dict(arrowstyle='->', color='#ef3b2c', lw=1.5),
    fontsize=11,
    color='#7f0000'
 )

plt.title('Loi de Poisson : probabilité du nombre d\'adoptions par semaine')
plt.xlabel('Nombre d\'adoptions en une semaine')
plt.ylabel('Probabilité P(X = k)')
plt.xticks(k_values)
plt.ylim(0, max(pmf_values) + 0.05)
plt.grid(axis='y', alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```
