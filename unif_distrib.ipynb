```python
from scipy.stats import uniform
```

```python
print(uniform.cdf(8, 0, 12)) # P(X <= 8)
print(1 - uniform.cdf(8, 0, 12)) # P(X > 8) = 1 - P(X <= 8)
```

Ci-dessus, proba d'attendre 8 min ou moins, 4min ou moins

```python
uniform.cdf(8, 0, 12) - uniform.cdf(4, 0, 12) # P(4 < X <= 8) = P(X <= 8) - P(X <= 4)
```

Entre 8 et 4 min

```python
uniform.rvs(0, 10, size=10) # Generate 10 random numbers from a uniform distribution between 0 and 10
```

```python
uniform.ppf(0.25, 0, 12) # Find the value x such that P(X <= x) = 0.25 for a uniform distribution between 0 and 12
```

```python
uniform.pmf(5, 0, 12) # For a continuous distribution like uniform, the PMF is 0 for any specific value.
```
